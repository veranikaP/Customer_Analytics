# Libraries

```python
import pandas as pd
import numpy as np
import scipy

import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

# Import Data

```python
df_segmentation = pd.read_csv('../segmentation data.csv', index_col = 0)
```

# Explore Data

```python
df_segmentation.head()
```

```python
df_segmentation.describe()
```

# Correlation Estimate

```python
df_segmentation.corr()
```

```python
plt.figure(figsize = (12, 9))
s = sns.heatmap(df_segmentation.corr(),
               annot = True,
               cmap = 'BrBG',
               vmin = -1,
               vmax = 1)
s.set_yticklabels(s.get_yticklabels(), rotation = 0, fontsize = 14)
s.set_xticklabels(s.get_xticklabels(), rotation = 90, fontsize = 14)
plt.title('Correlation Heatmap', fontsize = 14)
plt.show()
```

# Visualize Raw Data 

```python
plt.figure(figsize = (12, 9))
plt.scatter(df_segmentation.iloc[:, 2], df_segmentation.iloc[:, 4])
plt.xlabel('Age', fontsize = 14)
plt.ylabel('Income', fontsize = 14)
plt.title('Vizualization of raw data', fontsize = 14)
plt.show()
```

# Standardization

```python
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df_segmentation)
```

# Hierarchical Clustering 

```python
hier_clust = linkage(segmentation_std, method = 'ward')
```

```python
plt.figure(figsize = (12, 9))
plt.title('Hierarchical Clustering Dendrogram', fontsize = 14)
plt.xlabel('Observations', fontsize = 14)
plt.ylabel('Distance', fontsize = 14)
dendrogram(hier_clust,
           truncate_mode = 'level',
           p = 5,
           show_leaf_counts = False,
           no_labels = True)
plt.show()
```

# K-means Clustering

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(segmentation_std)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (12, 9))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--' )
plt.xlabel('Number of clusters', fontsize = 14)
plt.ylabel('WCSS', fontsize = 14)
plt.title('K-means clustering', fontsize = 14)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans.fit(segmentation_std)
```

# Results

```python
df_segm_kmeans = df_segmentation.copy()
df_segm_kmeans['Segment K-means'] = kmeans.labels_
```

```python
df_segm_analysis = df_segm_kmeans.groupby(['Segment K-means']).mean()
df_segm_analysis
```

### Description:
#### Segment 0 - Well-off. 
People with the highest income and level of education.
#### Segment 1 - Fewer-Opportunities. 
People with low income from small cities and low level of eduaction.
#### Segment 2 - Career-Focused. 
Men from middle sized and big cities. With enough high income and middle education. Single.
#### Segment 3 - Standard. 
People in relationships with average age of 29.

```python
df_segm_analysis['N Obs'] = df_segm_kmeans[['Segment K-means', 'Sex']].groupby(['Segment K-means']).count()
```

```python
df_segm_analysis['Prop Obs'] = df_segm_analysis['N Obs']/df_segm_analysis['N Obs'].sum()
```

```python
df_segm_analysis
```

```python
df_segm_analysis.rename({0: 'well-off',
                        1: 'fewer-opportunities',
                        2: 'career-focused',
                        3: 'standard'})
```

```python
df_segm_kmeans['Labels'] = df_segm_kmeans['Segment K-means'].map({0: 'well-off',
                                                                  1: 'fewer-opportunities',
                                                                  2: 'career-focused',
                                                                  3: 'standard'})
```

```python
x_axis = df_segm_kmeans['Age']
y_axis = df_segm_kmeans['Income']
plt.figure(figsize = (12, 9))
sns.scatterplot(x_axis, y_axis, hue = df_segm_kmeans['Labels'], palette = ['g', 'r', 'c', 'm'])
plt.title('Segmenation K-means')
plt.show()
```

# PCA

```python
pca = PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (12, 9))
plt.plot(range(1, 8), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explaied Variance by Components', fontsize = 14)
plt.xlabel('Number of Components', fontsize = 14)
plt.ylabel('Cumulative Explained Variance', fontsize = 14)
plt.show()
```

#### Rule of thumb
keep 80% of variance (from 80% - 90%)

```python
pca = PCA(n_components = 3)
```

```python
pca.fit(segmentation_std)
```

# PCA Results

```python
pca.components_
```

```python
df_pca_comp = pd.DataFrame(data = pca.components_,
                          columns = df_segmentation.columns.values,
                          index = ['Component 1', 'Component 2', 'Component 3'])
df_pca_comp
```

```python
plt.figure(figsize = (17, 5))
sns.heatmap(df_pca_comp,
           vmin = -1,
           vmax = 1, 
           cmap = 'BrBG',
           annot = True)
plt.yticks([0, 1, 2],
           ['Component 1', 'Component 2', 'Component 3'],
           rotation = 0,
           fontsize = 14)
plt.xticks(fontsize = 14)
plt.show()
```

```python
pca.transform(segmentation_std)
```

```python
scores_pca = pca.transform(segmentation_std)
```

# K-means clustering with PCA

```python
wcss = []
for i in range(1, 11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (12, 9))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--' )
plt.xlabel('Number of clusters', fontsize = 14)
plt.ylabel('WCSS', fontsize = 14)
plt.title('K-means with PCA clustering', fontsize = 14)
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans_pca.fit(scores_pca)
```

# K-means clustering with PCA Results

```python
df_segm_pca_kmeans = pd.concat([df_segmentation.reset_index(drop = True), pd.DataFrame(scores_pca)], axis=1)
df_segm_pca_kmeans.columns.values[-3:] = ['Component 1', 'Component 2', 'Component 3']
df_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans
```

```python
df_segm_pca_kmeans_freq
```
